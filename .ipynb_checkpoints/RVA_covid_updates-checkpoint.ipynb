```python
import pandas as pd
import datetime as dt
```

```python
#read in Department of Health Data and build list of dates and cities
url="https://www.vdh.virginia.gov/content/uploads/sites/182/2020/05/VDH-COVID-19-PublicUseDataset-Cases.csv"
df=pd.read_csv(url)
dates=df["Report Date"].unique()
df.head()
```

```python
rva_df=df.loc[df["Locality"]=="Richmond City",:]
henrico_df=df.loc[df["Locality"]=="Henrico",:]
chesterfield_df=df.loc[df["Locality"]=="Chesterfield",:]
pow_df=df.loc[df["Locality"]=="Powhatan",:]
pow_df.head()
```

```python
total_cases_r=rva_df['Total Cases'].iloc[0]
new_cases_r = total_cases_r-rva_df['Total Cases'].iloc[1]
total_hosp_r=rva_df['Hospitalizations'].iloc[0]
new_hosp_r = total_hosp_r-rva_df['Hospitalizations'].iloc[1]
new_death_r=rva_df['Deaths'].iloc[0]-rva_df['Deaths'].iloc[1]


total_cases_h=henrico_df['Total Cases'].iloc[0]
new_cases_h = total_cases_h-henrico_df['Total Cases'].iloc[1]
new_death_h=henrico_df['Deaths'].iloc[0]-henrico_df['Deaths'].iloc[1]
new_hosp_h = henrico_df['Hospitalizations'].iloc[0]-henrico_df['Hospitalizations'].iloc[1]

total_cases_c=chesterfield_df['Total Cases'].iloc[0]
new_cases_c = total_cases_c-chesterfield_df['Total Cases'].iloc[1]
new_death_c=chesterfield_df['Deaths'].iloc[0]-chesterfield_df['Deaths'].iloc[1]
new_hosp_c = chesterfield_df['Hospitalizations'].iloc[0]-chesterfield_df['Hospitalizations'].iloc[1]

total_cases_p=pow_df['Total Cases'].iloc[0]
new_cases_p = total_cases_p-pow_df['Total Cases'].iloc[1]
new_death_p=pow_df['Deaths'].iloc[0]-pow_df['Deaths'].iloc[1]
new_hosp_p = pow_df['Hospitalizations'].iloc[0]-pow_df['Hospitalizations'].iloc[1]
```

```python
print(f"Richmond had {new_cases_r} new cases of covid yesterday, bringing the total to {total_cases_r}" )
print(f"Henrico had {new_cases_h} new cases of covid yesterday, bringing the total to {total_cases_h}" )
print(f"Chesterfield had {new_cases_c} new cases of covid yesterday, bringing the total to {total_cases_c}" )
print(f"Powhatan had {new_cases_p} new cases of covid yesterday, bringing the total to {total_cases_p}" )

print(f"In Richmond, {new_hosp_r} people went to hospital and {new_death_r} people died from covid ")
print(f"In Henrico, {new_hosp_h} people went to hospital and {new_death_h} people died from covid ")
print(f"In Chesterfield, {new_hosp_c} people went to hospital and {new_death_c} people died from covid ")
print(f"In Powhatan, {new_hosp_p} people went to hospital and {new_death_p} people died from covid ")
```
